```python
class Analysis:
    def __init__(self, tree, option):
        self.tree = tree
        self.option = option
        self.accumulated_entries = 0
        

    def begin(self):
        nentries = self.tree.GetEntries()
        print("Processing ",nentries," entries in the "+self.option+" chain")
        self.Tmass = ROOT.TH1F("Tmass_"+self.option,"Invariant mass of W Boson",40,39,180)
    
    def process(self):
        
        for event in self.tree:
            self.accumulated_entries += 1
            
            #check for missing energy >30GeV
            if self.tree.met_et>30000:
                
                #Check for single electron or muon trigger
                if (self.tree.trigM) or (self.tree.trigE):
                    leptemp = ROOT.TLorentzVector()
                    nelectrons = 0
                    nmuons = 0
                    muonindex = -1
                    Wmass = 0
                    
                    #check that there is exactly 1 lepton with pt>35GeV
                    for ilepton in range(self.tree.lep_n):
                        leptemp.SetPtEtaPhiE(self.tree.lep_pt[ilepton], self.tree.lep_eta[ilepton], \
                                              self.tree.lep_phi[ilepton], self.tree.lep_E[ilepton])
                        
                        #if self.tree.lep_isTightID[ilepton]: commented out; this was here but within this if was
                        #moved 1 tab left for this edit
                            
                        #standard object selection criteria
                        if (self.tree.lep_pt[ilepton]/1000)>35 and \
                        ((self.tree.lep_ptcone30[ilepton]/self.tree.lep_pt[ilepton])<.1) and \
                        ((self.tree.lep_etcone20[ilepton]/self.tree.lep_pt[ilepton])<.1):


                            #electron selection
                            if self.tree.lep_type[ilepton] == 11 and \
                            TMath.Abs(self.tree.lep_eta[ilepton]) < 2.47 and \
                            ((TMath.Abs(self.tree.lep_eta[ilepton])<1.37) or (TMath.Abs(self.tree.lep_eta[ilepton])>1.52)):

                                if (TMath.Abs(self.tree.lep_trackd0pvunbiased[ilepton])/self.tree.lep_tracksigd0pvunbiased[ilepton])>5:
                                    continue
                                if (TMath.Abs(self.tree.lep_z0[ilepton]*math.sin(leptemp.Theta())) > .5):
                                    continue

                                nelectrons += 1

                            #muon selection
                            if self.tree.lep_type[ilepton] == 13 and (TMath.Abs(self.tree.lep_eta[ilepton]) < 2.5):

                                if (TMath.Abs(self.tree.lep_trackd0pvunbiased[ilepton])/self.tree.lep_tracksigd0pvunbiased[ilepton])>3:
                                    continue
                                if (TMath.Abs(self.tree.lep_z0[ilepton]*math.sin(leptemp.Theta())) > .5):
                                    continue

                                nmuons += 1
                                muonindex = ilepton
                                                                    
                                    
                    #make sure that lepton is a muon
                    if (nelectrons == 0) and (nmuons == 1):
                        Mvector = ROOT.TLorentzVector()
                        Nvector = ROOT.TLorentzVector()
                        weight = 0
                        
                        #Set muon and neutrino Lorentz vectors
                        Mvector.SetPtEtaPhiE (self.tree.lep_pt[muonindex], self.tree.lep_eta[muonindex], \
                                              self.tree.lep_phi[muonindex], self.tree.lep_E[muonindex])
                        
                        Nvector.SetPtEtaPhiE (self.tree.met_et, 0, self.tree.met_phi, self.tree.met_et)
                        
                        
                        #calculate W boson mass
                        Wmass = math.sqrt(2*Mvector.Pt()*Nvector.Et()*(1-math.cos(Mvector.DeltaPhi(Nvector))))
                        
                        
                        
                        #fill histograms
                        if Wmass >60000:
                            if self.option == "data":
                                weight = 1

                            if self.option != "data":
                                weight = (self.tree.mcWeight * self.tree.scaleFactor_PILEUP *\
                                          self.tree.scaleFactor_LepTRIGGER * self.tree.scaleFactor_MUON * \
                                          self.tree.scaleFactor_ELE)
                                
                            
                            self.Tmass.Fill(Wmass/1000, weight)

                        
                        
            if (self.accumulated_entries%1000000 == 1):
                print("Processed: ",self.accumulated_entries)

            
    def terminate(self):
        
        output_name = "output/WlvAnalysis_"+self.option+".root"
        print("Saving to file "+output_name)
        file = ROOT.TFile(output_name, "RECREATE")
        self.Tmass.Write()
        file.Close()
        print("Done saving!")
           
```
